```python
import pandas as pd
import numpy as np

def get_geom(file):
    atomos, geom = [], np.array([[0,0,0]])
    with open('geometries/'+file+'.xyz', 'r') as f:
        for line in f:
            line = line.split()
            if len(line) == 4:
                atomos.append(line[0])
                coord = np.array([[float(line[1]),float(line[2]),float(line[3])]])
                geom  = np.vstack((geom,coord))
    return atomos, geom[1:,:]
```

```python
data = pd.read_csv('database.csv')
data
```

```python
np.set_printoptions(precision=4,suppress=True)
molecules = data['molecule'].values
at, ge = get_geom(molecules[0])    
at,ge
```

```python
def print_geom(file):
    with open('geometries/'+file+'.xyz','r') as f:
        text = f.read()
    print(text)

print_geom(molecules[0])
```
