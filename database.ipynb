```python
import pandas as pd
import numpy as np
np.set_printoptions(precision=4,suppress=True)


data = pd.read_csv('database.csv')
data
```

```python
def print_geom(file):
    with open('geometries/'+file+'.xyz','r') as f:
        text = f.read()
    return text[4:]

a =print_geom(data['molecule'][0])
print(a)
```

```python
molecules = data['molecule'].values
for mol in molecules:
    geom = print_geom(mol)
    with open('geom.com', 'r') as f:
        text = f.read().replace('AAAA',geom[4:])
    with open(f'{mol}.com','w') as g:
        g.write(text+'\n\n')
```
